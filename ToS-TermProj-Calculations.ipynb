```python
from anastruct import SystemElements
ss = SystemElements()
```

```python
ss.add_element(location=[[0,0],[0,3]])
ss.add_element(location=[[0,3],[1.75,3]])
ss.add_element(location=[[1.75,3],[3.5,3]])
ss.add_element(location=[[0,0],[1.75,0]])
ss.add_element(location=[[1.75,0],[3.5,0]])
ss.add_element(location=[[0,3],[1.75,0]])
ss.add_element(location=[[1.75,0],[3.5,3]])
ss.add_element(location=[[3.5,3],[3.5,0]])
ss.add_element(location=[[1.75,3],[1.75,0]])

ss.add_support_hinged(node_id=2)
ss.add_support_roll(node_id=1, direction=1)
```

```python
ss.q_load(q=-2.37, element_id=2)
ss.q_load(q=-2.37, element_id=3)
```

```python
ss.show_structure()
```

```python
ss.solve()
ss.show_reaction_force()
ss.show_bending_moment()
ss.show_shear_force(verbosity=0)
ss.show_axial_force()
```
